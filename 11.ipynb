## 02. hist()와 boxplot() 그리기

1. 히스토그램

- 데이터의 분포 상태를 직사각형 모양의 도수 분포 그래프를 그려줌
  - 데이터의 빈도에 따라 직사각형의 높이(빈도 수)가 결정됨
  - hist() 함수 사용

```python
import matplotlib.pyplot as plt

plt.hist([1, 1, 2, 3, 4, 5, 6, 6, 7, 8, 10]) # 각각의 숫자가 존재하는 빈도를 표로 그려줌
plt.show()
```

- 자료에 표시된 최소값에서부터 최대값까지 10개의 동일한 구간으로 나누어 각각의 구간의 빈도를 표시
- 함수 hist()의 반환값에 모두 저장
  - count: 각 구간의 빈도 수 배열
  - Bins: 각 구간을 나누는 수 배열

`[1., 1.9) [1.9, 2.8) ... [8.2, 9.1), [9.1, 10.]`

```python
import matplotlib.pyplot as plt

count, bins, c = plt.hist([1, 1, 2, 3, 4, 5, 6, 6, 7, 8, 10])
plt.show()
print(count) # 각 구간에 있는 빈도 수 
print(bins) 
```

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

dice = [rd.randint(1, 6) for i in range(1000000)]
plt.hist(dice, bins=np.arange(0.5, 7), rwidth=.9)
plt.xticks(range(1, 7))
plt.show()

# 주사위를 던지는 횟수가 늘어날 수록 주사위의 특정 숫자가 나오는 횟수가 전체의 1/6에 가까워 짐을 알 수 있다. (큰수의 법칙)
```

### 상자 그림(Boxplot)

- 데이터에서 얻어낸 최대값, 최소값, 상위 1/4, 2/4(중앙), 3/4에 위치한 값을 보여주는 그래프

```python
import matplotlib.pyplot as plt
import random as rd

result = [rd.randint(1, 1000) for i in range(13)]
print(sorted(result))
plt.boxplot(result)
plt.show()
```

#### 하나의 그래프에서 두 데이터의 상자 그림 그리기(1)

```python
import matplotlib.pyplot as plt
import random as rd

result1 = []
result2 = []

for i in range(13):
    result1.append(rd.randint(1, 500))
    result2.append(rd.randint(501, 1000))

plt.boxplot(result1)
plt.boxplot(result2)
plt.show()
```

#### 하나의 그래프에서 두 데이터의 상자 그림 그리기(2)

```python
import matplotlib.pyplot as plt
import random as rd

result1 = []
result2 = []

for i in range(13):
    result1.append(rd.randint(1, 500))
    result2.append(rd.randint(501, 1000))

plt.boxplot([result1, result2])
plt.show()
```
