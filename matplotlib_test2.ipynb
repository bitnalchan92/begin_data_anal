```python
import matplotlib.pyplot as plt
```

## 그래프에 제목(title) 추가하기

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40]) # x축의 데이터와 y축의 데이터 개수가 같지 않으면 오류가 발생
plt.title('plotting')
plt.show()
```

## 그래프에 x축 및 y축 레이블(Label) 정보 추가하기

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40]) # x축의 데이터와 y축의 데이터 개수가 같지 않으면 오류가 발생
plt.title('plotting')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

## 그래프에 범례(Legend) 추가하기

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], label='ascending')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], label='descending')

plt.title('Legend')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.show()
```

## 색상과 스타일, 축의 최대 최소 지정

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis((0, 6, 0, 20)) # plt.axis = [xmin, xmax, ymin, ymax]

plt.show()
```

## 그래프의 색상 바꾸기
- red -> r
- green -> g
- blue -> b
- black -> k
- yellow -> y
- white -> w
- magenta -> m
- cyan -> c

> http://matplotlib.org/examples/color/named_colors.html

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='skyblue', label='skyblue', ls='--')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'pink', label='pink', linestyle='-.')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

### 마커 marker

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='skyblue', label='diamond', ls='--', marker='d')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'pink', label='triangle up', linestyle='-.', marker='^')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

### 선의 굵기 바꾸기

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='skyblue', linewidth=3, label='thick line', ls='--', marker='d')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'pink', linewidth=0.5, label='thin line', linestyle='-.', marker='^')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

### 문자열 형태로 라인 모양
`fmt = '[color][marker][line]`

- 'b' : blue markers with default shape
- 'or' : red circles
- '-g' : green solid line
- '--' : dashed line with default color
- '^k:' : black triangle_up markers connected by a dotted line

```python
# 라인스타일이 없으면 마커만 찍힘
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], 'r.', label='circle') # 빨간색 원형 마커 그래프
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'g^', label='triangle up') # 초록색 삼각형 마커 그래프

plt.title('marker') # 제목 설정
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend() # 범례 표시
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], '-.m*', [1, 2, 3, 4], [50, 60, 70, 80], 'r.', label='circle')
plt.plot([40, 30, 20, 10], 'g^', label='triangle up')

plt.title('marker') # 제목 설정
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend() # 범례 표시
plt.show()
```

```python
# mec 마크 엣지 컬러
# mew 마크 엣지 굵기
# mfc 마크 안의 컬러
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='b',
         lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')

plt.title('Style')
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], "-.bo", lw=3, ms=8, mec="skyblue", mew=2, mfc="y")
plt.title("Style")
plt.show()
```
