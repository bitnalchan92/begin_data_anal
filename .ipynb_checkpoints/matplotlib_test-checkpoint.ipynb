## plt.plot()
- 선 그래프 그리기

```python
import matplotlib.pyplot as plt

plt.plot([0, 1, 2, 3]) # y축에 해당하는 값, list를 하나만 대입할시 x축은 자동으로 0, 1, 2, 3이 대응된다.
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40]) # y축에 해당하는 값, list를 하나만 대입할시 x축은 자동으로 0, 1, 2, 3이 대응된다.
plt.show()
```

```python
import matplotlib.pyplot as plt

# plot([x축 데이터], [y축 데이터])
plt.plot([1, 2, 3, 4], [10, 20, 30, 40])
plt.show()
```
