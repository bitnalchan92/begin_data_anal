```python
import matplotlib.pyplot as plt

#         x축 데이터      y축 데이터 
plt.plot([1, 2, 3, 4], [10, 20, 30, 40]) # x축 데이터와 y축 데이터의 개수가 같지 않으면 오류가 발생한다.
plt.title('plotting')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40])
plt.title('plotting')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40], label='ascending')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], label='descending')
plt.title('legend')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend() # 이 문장을 사용해야 범례가 출력된다.
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # r은 color, o는 circle모양을 명시함
plt.axis((0, 6, 0, 20))
plt.show()
```

# 그래프의 색상 바꾸기

- red -> r
- green -> g
- blue -> b
- black -> k
- yellow -> y
- white -> w
- magenta -> m
- cyan -> c

> http://matplotlib.org/examples/color/named_colors.html

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='skyblue', label='skyblue')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'pink', label='pink')
plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

# 그래프의 선 모양 바꾸기

- `-` : solid line
- `--` : dashed line
- `-.` : dash-dot line
- `:` : dotted line

## 마커 marker
- 데이터의 위치를 나타내는 기호 ( 실제 데이터를 돋보이도록... )

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='r', linestyle='--', marker='D', linewidth=3, label='dashed')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'b', ls=':', marker='^', label='dotted')
plt.title('linestyle')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```
